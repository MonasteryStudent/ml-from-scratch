# Linear Regression Demo

This notebook demonstrates univariate linear regression using gradien descent with a simple application scenario.

The goal is to fit a line to a small synthetic dataset with a linear relationship. The features of the dataset are made up exam experiences of fellow students at universtity. We collected data about the relationsship between study hours (1-20) and exam scores (20-100).

```python
import numpy as np
import matplotlib.pyplot as plt

from ml_from_scratch.linear_regression import uni_linear_regression, predict
```

```python
np.random.seed(42)
hours = np.arange(1, 21)
# To introduce some noise we use the normal distribution
score = 4 * hours + 20 + np.random.normal(0, 10, len(hours))
```

```python
plt.scatter(hours, score)
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Training Data")
plt.show()
```

In the plot above you see a cloud of datapoints. It seems that the we have a linear relationship in our data, we could try and draw a line per hand through the data. Like this our line would be more or less optimal, so a better way is to use linear regression. For that we will train our linear regression model in the code cell below. 

```python
w, b, _ = uni_linear_regression(
    hours,
    score,
    alpha=0.01,
    iterations=5000
)
```

As you can see the function returns w and b: Those parameters are the slope w and the intercept b of a linear equation $f_{w, b}(x) = wx + b$. This linear equation is also implemented in the predict function, that we use in the code cell below to calculate predicted exam scores based on our study hours training data.

```python
score_pred = predict(hours, w, b)
```

```python
plt.scatter(hours, score, label="Training data")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.plot(hours, score_pred, label="Linear regression")
plt.legend()
plt.show()
```

As you can see in the plot above the line we draw with our predicted scores fits pretty well into our data distribution and should therefor be useful to make predictions. Say we want to know what score we can expect if we would learn 18 h for our next exam at university.

```python
prediction_hours = 18 
prediction_score = predict(prediction_hours, w, b)
```

```python
plt.scatter(hours, score, label="Training data")
plt.plot(hours, score_pred, label="Linear regression")

# vertical line
plt.plot(
    [prediction_hours, prediction_hours],
    [0, prediction_score],
    "r--"
)

# horizontal line
plt.plot(
    [0, prediction_hours],
    [prediction_score, prediction_score],
    "r--"
)

# prediction point
plt.scatter(
    prediction_hours,
    prediction_score,
    label="Prediction"
)

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

As you can see in the plot, we draw a vertical line from 18 h on the x-axis towards our regression line and read the corresponding exam score on the y-axis. The prediction tells us that we can expect around 82 points.
